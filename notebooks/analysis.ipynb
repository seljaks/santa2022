```python
slow_group = stats.groupby('slow_counter')

slow_agg = slow_group.agg(Sum=('step_cost', 'sum'),
               Mean=('step_cost', 'mean'),
               Count=('step_cost', 'count'),
               start_x=('x', 'first'),
               start_y=('y', 'first'),
               last_x=('x', 'last'),
               last_y=('y', 'last'),
               last_dx=('dx', 'last'),
               last_dy=('dy', 'last'),
               last_index=('index', 'last')
               ).sort_values('Sum', ascending=False)

slow_agg['last_index'] = slow_agg['last_index'] + 1
slow_agg['end_x'] = slow_agg['last_x'] + slow_agg['last_dx']
slow_agg['end_y'] = slow_agg['last_y'] + slow_agg['last_dy']
slow_agg = slow_agg.drop(columns=['last_x', 'last_y', 'last_dx', 'last_dy'])
slow_agg
```
